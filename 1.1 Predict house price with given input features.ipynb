# Predicting house prices with given input features

### We will use scikit-learn to run the models

```python
import sklearn
from sklearn.model_selection import train_test_split
import pandas as pd
```

Here, we imported scikit-learn (sklearn) and pandas libraries.

Pandas is an open source library providing high-performance, easy-to-use data structures and data analysis tools for the Python programming. In particular, it offers data structures and operations for manipulating numerical tables.

##### Load the dataset 

```python
sales = pd.read_csv('sales.csv')
```

##### Let's look at the dataset

```python
sales
```

Let's import matplotlib.pyplot and make the output of the graphs to be in the notebook itself.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Finally lets see the correlation between 'Living area in sqft' with the 'Price'.

```python
plt.scatter(sales['sqft_living'], sales['price'])
plt.show()
```

Next, we create a simple regression model based on the above graph to predict house price based on only the 'sqft_living' feature.

We split the sales data into train and test data in the ratio 4:1.

```python
train_data = sales.iloc[0: int(8 * len(sales)/10)]
test_data = sales.iloc[int(8 * len(sales)/10):]
features = ['sqft_living']
X = train_data[features]
y = train_data['price']
```

We import the Linear Regression model from the sklearn library.

```python
from sklearn.linear_model import LinearRegression
```

We build the model using sklearn's built in linear regression model.

```python
sqft_model = LinearRegression(fit_intercept =True)
```

Now we train our model by passing the features (sqft_living as X) along with the target (price as y).

```python
sqft_model.fit(X,y)
```

The model calculates the optimal value of the coefficients (w<sub>0</sub>, w<sub>1</sub>) which minimises the RSS.

Actually, the parameters returned are the optimal parameters. So according to the convention we introduced in the slides, the above w<sub>0</sub> and w<sub>1</sub> are actually w<sub>0</sub> <b>cap</b> and w<sub>1</sub> <b>cap</b>.

But we are unable to write the cap signs in Markdown. If you know how to write the symbol, create a Pull Request. We will merge it.


Now let's view the model coefficients. First, we get the y-intercept, followed by the slope.

```python
print (sqft_model.intercept_)
print (sqft_model.coef_)
```

Next, we evaluate on the <b>test data</b> to find the errors of this model.

<b>rmse</b> - Root mean square error of all the points in the test dataset

```python
from sklearn.metrics import mean_squared_error
rmse_sqft_model = mean_squared_error(test_data['price'], sqft_model.predict(test_data[features]))**0.5
```

```python
print (rmse_sqft_model)
```

Now, we plot our linear regression model.

```python
plt.plot(test_data['sqft_living'],test_data['price'], '.') 
plt.plot(test_data['sqft_living'],sqft_model.predict(test_data[features].values.reshape(4323,1)), '-')
plt.show()
```

Blue dots are original data, orange line is the prediction from the simple regression.

This was just a simple regression model using sqft_living as input and price as output.

### Lets use more features as input and create a multiple regression model.

```python
my_model = LinearRegression(fit_intercept =True)
```

```python
my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
```

```python
X = train_data[my_features]
```

```python
my_model.fit(X,y)
```

```python
print (my_model.intercept_)
print (my_model.coef_)
```

```python
rmse_my_model = mean_squared_error(test_data['price'], my_model.predict(test_data[my_features]))**0.5
```

```python
print (rmse_my_model)
```

### Lets create another advanced model using much more features as input

```python
advanced_model = LinearRegression(fit_intercept =True)
```

```python
advanced_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode', 'condition', 'grade',
                    'waterfront', 'view', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long',
                    'sqft_living15', 'sqft_lot15']
```

```python
X = train_data[advanced_features]
```

```python
advanced_model.fit(X,y)
```

```python
print (advanced_model.intercept_)
print (advanced_model.coef_)
```

```python
rmse_advanced_model = mean_squared_error(test_data['price'], advanced_model.predict(test_data[advanced_features]))**0.5
```

```python
print rmse_advanced_model
```

On comparing the three models, we get minimum RMSE on the advanced_model. This means that for this case study, adding more features increases the model quality.

###### The rmse decreases with the increase in complexity of the model.
